```python
merged_months = pd.concat([renamed_jan, renamed_feb, renamed_mar, renamed_apr, renamed_may, renamed_jun,
                       renamed_jul, renamed_aug, renamed_sept, renamed_octo, renamed_nov, renamed_dec])
merged_months['Start Time'] = merged_months['Start Time'].apply(lambda x:x.replace(minute=0, second=0, microsecond=0))
merged_months.head()
```

```python
user ={
   "Subscriber" : 1,
   "Customer" : 2}
stype = {"Subscriber" : 1,
      "Customer": 0}
ctype = {"Subscriber" : 0,
      "Customer": 1}
merged_months['binary_type'] = merged_months['User Type'].map(user)
merged_months['Subscribers'] = merged_months['User Type'].map(stype)
merged_months['Non Subscribers'] = merged_months['User Type'].map(ctype)
merged_months.head(20)
```

```python
hours_grouped = merged_months.groupby("Start Time").agg({'Trip Duration (minutes)':'mean',
                                            'User Type':'count', 'Subscribers': 'sum', 'Non Subscribers': 'sum'})
hours_regrouped = hours_grouped.rename(columns={'Trip Duration (minutes)': 'Average Trip Duration', 'User Type': 'Amount of Rides'})
hours_regrouped['Subscriber %'] = round(hours_regrouped['Subscribers'] / hours_regrouped['Amount of Rides'] * 100, 0)
hours_regrouped['Non Subscriber %'] = round(hours_regrouped['Non Subscribers'] / hours_regrouped['Amount of Rides'] * 100, 0)
hours_final = hours_regrouped.reset_index()
hours_final.info()
```

```python
hours_final.to_csv("FinalBostonMerged.csv", index=False, header=True)
```
