```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Resources of CSV files/MonthlyWeatherBostonMerged.csv")
df2 = pd.read_csv("Resources of CSV files/MonthlyWeatherPortlandMerged.csv")
```

```python
df1.head()
```

```python
dateindex=df1['dt_iso']
x = df1['dt_iso']
aa=dict(zip(x,df1['Amount of Rides']))
bb=dict(zip(x,df2['Amount of Rides']))
bars = ('Jan', 'Feb', 'Mar', 'Apr', 'Mar', 'Jun', 'Jul','Aug','Sep','Oct','Nov','Dec')


dfbar = pd.DataFrame({'Boston': aa, 'Portland': bb}, index=dateindex)

# Non-stacked bar plot
dfbar.plot.bar(figsize=(10, 9))

plt.title("Rides per Month")
plt.legend(title="City")
plt.xlabel('Months')
plt.ylabel('Rides')
plt.xticks(np.arange(12), bars)
plt.show() 
```
