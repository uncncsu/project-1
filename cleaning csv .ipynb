```python
jan = "Resources/201801_hubway_tripdata.csv"
feb = "Resources/201802_hubway_tripdata.csv"
march = "Resources/201803_hubway_tripdata.csv"
april = "Resources/201804_hubway_tripdata.csv"
may = "Resources/201805-bluebikes-tripdata.csv"
june = "Resources/201806-bluebikes-tripdata.csv"
july = "Resources/201807-bluebikes-tripdata.csv"
aug = "Resources/201808-bluebikes-tripdata.csv"
sep = "Resources/201809-bluebikes-tripdata.csv"
oc = "Resources/201810-bluebikes-tripdata.csv"
nov = "Resources/201811-bluebikes-tripdata.csv"
dec = "Resources/201812-bluebikes-tripdata.csv"
```

```python
jan_df = pd.read_csv(jan) 
new_jan = jan_df[['tripduration', 'starttime', 'usertype']]
final_jan = new_jan.dropna()
renamed_jan = final_jan.rename(columns={'tripduration': 'Trip Duration', 'starttime': 'Start Time', 
                                        'usertype': 'User Type'})
renamed_jan['Trip Duration'] = renamed_jan['Trip Duration']/60


renamed_jan.head()
```

```python
feb_df = pd.read_csv(feb)
new_feb = feb_df[['tripduration', 'starttime', 
                        'usertype']]
final_feb = new_feb.dropna()
renamed_feb = final_feb.rename(columns={'tripduration': 'Trip Duration', 
                                        'starttime': 'Start Time', 'usertype': 'User Type'})


renamed_feb['Start Time'] = pd.to_datetime(renamed_feb['Start Time'])
renamed_feb['Start Time'] = renamed_feb['Start Time'].dt.round('1s')

renamed_feb.head()
```
