```python
import pandas as pd
```

```python
jan = "Resources/201801_hubway_tripdata.csv"
feb = "Resources/201802_hubway_tripdata.csv"
march = "Resources/2018_03.csv"
april = "Resources/2018_04.csv"
may = "Resources/2018_05.csv"
june = "Resources/2018_06.csv"
july = "Resources/2018_07.csv"
aug = "Resources/2018_08.csv"
sep = "Resources/2018_09.csv"
oc = "Resources/2018_10.csv"
nov = "Resources/2018_11.csv"
dec = "Resources/2018_12.csv"
```

```python
jan_df = pd.read_csv(jan) 
new_jan = jan_df[['tripduration', 'starttime', 
                        'stoptime', 'usertype', 
                        'birth year', 'gender']]
final_jan = new_jan.dropna()
renamed_jan = final_jan.rename(columns={'tripduration': 'Trip Duration', 'starttime': 'Start Time', 
                        'stoptime': 'Stop Time', 'usertype': 'User Type', 
                        'birth year': 'Birth Year', 'gender': 'Gender'})
renamed_jan['Age'] = 2018 - (renamed_jan['Birth Year'])
renamed_jan['Gender'] = renamed_jan['Gender'].replace([0,1,2],['Female','Male','Other'])
renamed_jan
```

```python
feb_df = pd.read_csv(feb)
new_feb = feb_df[['tripduration', 'starttime', 
                        'stoptime', 'usertype', 
                        'birth year']]
final_feb = new_feb.dropna()
renamed_feb = final_feb.rename(columns={'tripduration': 'Trip Duration', 'starttime': 'Start Time', 
                        'stoptime': 'Stop Time', 'usertype': 'User Type', 
                        'birth year': 'Age'})
renamed_feb['Age'] = 2018 - (renamed_feb['Age'])

new = renamed_feb["Start Time"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
renamed_feb["Start Date"]= new[0] 
  
# making separate last name column from new data frame 
renamed_feb["Time Start"]= new[1] 
  
# Dropping old Name columns 
renamed_feb.drop(columns =["Start Time"], inplace = True) 
renamed_feb
```
