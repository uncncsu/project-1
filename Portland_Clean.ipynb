```python
jan_df = pd.read_csv(jan)

new_jan = jan_df[['StartDate', 'StartTime', 'EndTime', 'Duration', 'PaymentPlan']]

newer_jan = new_jan.dropna()

newerer_jan = newer_jan.rename(columns={'Duration': 'Trip Duration', 
                                        'StartDate': 'Start D',
                                        'StartTime': 'Start T',
                                        'EndTime': 'End Time',
                                        'PaymentPlan': 'User Type',})

newerer_jan['Start Time'] = pd.to_datetime(newerer_jan['Start D'] + ' ' + newerer_jan['Start T'])
# newerer_jan['End Time'] = pd.to_datetime(newerer_jan['End Time']).dt.time
# newerer_jan['Start Date'] = pd.to_datetime(newerer_jan['Start Date']).dt.date
newest_jan = newerer_jan.drop(columns=['Start D', 'Start T'])
newnew_jan = newest_jan[['Start Time', 'Trip Duration', 'User Type']]
#newnew_jan['Trip Duration'] = pd.to_datetime(newnew_jan['Trip Duration'])
def convert_to_minutes(value):
    hours, minutes = value.split(':')
    return int(hours) * 60 + int(minutes)
newnew_jan['Trip Duration'] = newnew_jan['Trip Duration'].apply(convert_to_minutes)

newnew_jan.head(50)
```
