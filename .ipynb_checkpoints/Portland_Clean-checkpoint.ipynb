```python
import pandas as pd
```

```python
jan = "Resources/2018_01.csv"
feb = "Resources/2018_02.csv"
mar = "Resources/2018_03.csv"
apr = "Resources/2018_04.csv"
may = "Resources/2018_05.csv"
jun = "Resources/2018_06.csv"
jul = "Resources/2018_07.csv"
aug = "Resources/2018_08.csv"
sep = "Resources/2018_09.csv"
oct = "Resources/2018_10.csv"
nov = "Resources/2018_11.csv"
dec = "Resources/2018_12.csv"

months = [jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec]
```

```python
jan_df = pd.read_csv(jan)

new_jan = jan_df[['starttime','tripduration', 'usertype', 'birth year']]

newer_jan = new_jan.dropna()

newerer_jan = newer_jan.rename(columns={'tripduration': 'Trip Duration',
                                        'starttime': 'Start Time',
                                        'usertype': 'User Type',
                                        'birth year': 'Birth Year'})
newerer_jan['Age'] = 2018 - (newerer_jan['Birth Year'])

newest_jan = newerer_jan.drop(columns=['Birth Year'])
newest_jan['Start Time'] = pd.to_datetime(newest_jan['Start Time'])
newest_jan['Start Time'] = newest_jan['Start Time'].dt.round('1s')
newest_jan['Start Time'] = newest_jan['Start Time'].dt.time
#newest_jan.head()
```
