```python
jan_df = pd.read_csv(jan)

new_jan = jan_df[['StartDate', 'StartTime', 'Duration', 'PaymentPlan']]

newer_jan = new_jan.dropna()

newerer_jan = newer_jan.rename(columns={'Duration': 'Trip Duration', 
                                        'StartDate': 'Start D',
                                        'StartTime': 'Start T',
                                        'PaymentPlan': 'User Type',})

newerer_jan['Start Time'] = pd.to_datetime(newerer_jan['Start D'] + ' ' + newerer_jan['Start T'])
newest_jan = newerer_jan.drop(columns=['Start D', 'Start T'])
newnew_jan = newest_jan[['Trip Duration', 'Start Time', 'User Type']]

def convert_to_minutes(value):
    hours, minutes = value.split(':')
    return int(hours) * 60 + int(minutes)
newnew_jan['Trip Duration'] = newnew_jan['Trip Duration'].apply(convert_to_minutes)

newnew_jan.head()
```

```python
merged_months = pd.concat([newnew_jan, newnew_feb, newnew_mar, newnew_apr, 
                           newnew_may, newnew_jun, newnew_jul, newnew_aug,
                          newnew_sep, newnew_oct, newnew_nov, newnew_dec])
merged_months.head()
```

```python
merged_months['Start Time'] = merged_months['Start Time'].apply(lambda x:x.replace(minute=0))

merged_months.head(20)
```

```python
user ={
  "Subscriber" : 1,
  "Casual" : 2}
stype = {"Subscriber" : 1,
       "Casual": 0}
ctype = {"Subscriber" : 0,
       "Casual": 1}
merged_months['binary_type'] = merged_months['User Type'].map(user)
merged_months['Subscribers'] = merged_months['User Type'].map(stype)
merged_months['Non Subscribers'] = merged_months['User Type'].map(ctype)
merged_months.head(20)
```

```python
hours_grouped = merged_months.groupby("Start Time").agg({'Trip Duration':'mean',
                                             'User Type':'count', 'Subscribers': 'sum', 'Non Subscribers': 'sum'})
hours_regrouped = hours_grouped.rename(columns={'Trip Duration': 'Average Trip Duration', 'User Type': 'Amount of Rides'})
hours_regrouped['Subscriber %'] = round(hours_regrouped['Subscribers'] / hours_regrouped['Amount of Rides'] * 100, 0)
hours_regrouped['Non Subscriber %'] = round(hours_regrouped['Non Subscribers'] / hours_regrouped['Amount of Rides'] * 100, 0)
hours_regrouped.head(20)
```
