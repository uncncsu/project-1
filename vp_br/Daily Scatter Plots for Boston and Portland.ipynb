```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bos = "DailyWeatherBostonMerged.csv"
port = "DailyWeatherPortlandMerged.csv"
```

```python
bos_df = pd.read_csv(bos)
port_df = pd.read_csv(port)
```

```python
bos_df
```

```python
ax = plt.gca()

bos_df.plot(kind='scatter',x='avg_temp_f',y='Amount of Rides',ax=ax, color="r", legend='Boston')


plt.xlim(-10.0, max(bos_df["avg_temp_f"])+5)
plt.ylim(0, max(bos_df["Amount of Rides"])+1000)
plt.title("Rides vs. Average Temp")
#plt.legend(loc="best")

plt.show()

```
