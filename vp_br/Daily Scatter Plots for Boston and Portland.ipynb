```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
bos = "DailyWeatherBostonMerged.csv"
port = "DailyWeatherPortlandMerged.csv"
```

```python
bos_df = pd.read_csv(bos)
port_df = pd.read_csv(port)
```

```python
bos_df
```

```python
ax = plt.gca()

bos_df.plot(kind='scatter',x='avg_temp_f',y='Amount of Rides',ax=ax, color="r")
port_df.plot(kind='scatter',x='avg_temp_f',y='Amount of Rides',ax=ax)



plt.xlim(-10.0, max(bos_df["avg_temp_f"])+5)
plt.ylim(0, max(bos_df["Amount of Rides"])+1000)
plt.title("Rides vs. Average Temp")
plt.xlabel("Average Temperature")
plt.ylabel("Amount of Rides")

#plt.legend(loc=2)
red_patch = mpatches.Patch(color='red', label='Boston')
blue_patch = mpatches.Patch(color='blue', label='Portland')

plt.legend(handles=[red_patch, blue_patch])

#b = mpatches.Patch(color='red', label='Boston')
#plt.legend(handles=[b], loc=2)


#p = mpatches.Patch(color='blue', label='Portland')
#plt.legend(handles=[p], loc=2)

plt.show()

```
