```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
p_df = "Final_Portland_Merged.csv"
b_df = "FinalBostonMerged.csv"
wb_df = "hour_weather_bos.csv"
wp_df = "hour_weather_por.csv"
```

```python
port_df = pd.read_csv(p_df)
bost_df = pd.read_csv(b_df)
w_b = pd.read_csv(wb_df)
w_p = pd.read_csv(wp_df)
```

```python
w_b.head()
```

```python
renamed_p = port_df.rename(columns={'Start Time': 'dt_iso' })
renamed_b = bost_df.rename(columns={'Start Time': 'dt_iso' })
```

```python
renamed_b.head()
```

```python
combined_df = pd.merge(w_b, renamed_b,
                                 on='dt_iso', how='left', right_index=True, left_index=True)
combined_df.head()
```

```python
combined_df.count()
```

```python
combined_df.columns
```

```python
bosw_df = combined_df.dropna(subset=['Average Trip Duration',
       'Amount of Rides', 'Subscribers', 'Non Subscribers', 'Subscriber %',
       'Non Subscriber %'])
```

```python
bosw_df.count()
```

```python
bosw_df['weather_type'].unique()
```

```python
cleanbosw_df = bosw_df.dropna(subset=['weather_type'])
cleanbosw_df.count()
```

```python
cleanbosw_df.head()
```

```python
cleanbosw_df.to_csv("WeatherBostonMerged.csv", index=False, header=True)
```

```python
combinedp_df = pd.merge(w_p, renamed_p,
                                 on='dt_iso', how='left', right_index=True, left_index=True)
combinedp_df.head()
```

```python
combinedp_df.count()
```

```python
porw_df = combinedp_df.dropna(subset=['Average Trip Duration',
       'Amount of Rides', 'Subscribers', 'Non Subscribers', 'Subscriber %',
       'Non Subscriber %'])
```

```python
porw_df.count()
```

```python
porw_df.to_csv("WeatherPortlandMerged.csv", index=False, header=True)
```
