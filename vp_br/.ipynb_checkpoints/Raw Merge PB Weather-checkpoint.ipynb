```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
```

```python
p_df = "Final_Portland_Merged.csv"
b_df = "FinalBostonMerged.csv"
wb_df = "hour_weather_bos.csv"
wp_df = "hour_weather_por.csv"
```

```python
port_df = pd.read_csv(p_df)
bost_df = pd.read_csv(b_df)
w_b = pd.read_csv(wb_df)
w_p = pd.read_csv(wp_df)
```

```python
w_b.head()
```

```python
renamed_p = port_df.rename(columns={'Start Time': 'dt_iso' })
renamed_b = bost_df.rename(columns={'Start Time': 'dt_iso' })
```

```python
renamed_b.head()
```

```python
combined_df = pd.merge(w_b, renamed_b,
                                 on='dt_iso', how='left', right_index=True, left_index=True)
combined_df.head()
```

```python
combined_df.count()
```

```python
combined_df.columns
```

```python
bosw_df = combined_df.dropna(subset=['Average Trip Duration',
       'Amount of Rides', 'Subscribers', 'Non Subscribers', 'Subscriber %',
       'Non Subscriber %'])
```

```python
bosw_df.count()
```

```python
bosw_df['weather_type'].unique()
```

```python
cleanbosw_df = bosw_df.dropna(subset=['weather_type'])
cleanbosw_df.count()
```

```python
cleanbosw_df.to_csv("HourlyWeatherBostonMerged.csv", index=False, header=True)
```

```python
cleanbosw_df['dt_iso'] = pd.to_datetime(cleanbosw_df['dt_iso'], format='%Y-%m')
cleanbosw_df['dt_iso'] = cleanbosw_df['dt_iso'].apply(lambda x:x.replace(hour=0))
cleanbosw_df.head()
cleanbosw_df.columns
```

```python
cleanbosw_df = cleanbosw_df.groupby('dt_iso').agg({'city_id': 'max',
                                   'pressure': 'mean',
                                   'humidity': 'mean',
                                   'wind_speed': 'mean',
                                   'avg_temp_f': 'mean',
                                   'weather_type': pd.Series.mode,
                                   'Average Trip Duration':'mean',
                                'Amount of Rides':'sum', 
                                   'Subscribers': 'sum', 
                                   'Non Subscribers': 'sum'})
cleanbosw_df['Subscriber %'] = round(cleanbosw_df['Subscribers'] / cleanbosw_df['Amount of Rides'] * 100, 0)
cleanbosw_df['Non Subscriber %'] = round(cleanbosw_df['Non Subscribers'] / cleanbosw_df['Amount of Rides'] * 100, 0)
day_bosw_df = cleanbosw_df.reset_index()
day_bosw_df.head()
```

```python
day_bosw_df.to_csv("DailyWeatherBostonMerged.csv", index=False, header=True)
```

```python
day_bosw_df.head()
```

```python
day_bosw_df['dt_iso'] = day_bosw_df['dt_iso'].astype(str).str[:-3]
```

```python
day_bosw_df.head()
```

```python
day_bosw_df = day_bosw_df.groupby('dt_iso').agg({'city_id': 'max',
                                   'pressure': 'mean',
                                   'humidity': 'mean',
                                   'wind_speed': 'mean',
                                   'avg_temp_f': 'mean',
                                   'weather_type': pd.Series.mode,
                                   'Average Trip Duration':'mean',
                                'Amount of Rides':'sum', 
                                   'Subscribers': 'sum', 
                                   'Non Subscribers': 'sum'})
day_bosw_df['Subscriber %'] = round(day_bosw_df['Subscribers'] / day_bosw_df['Amount of Rides'] * 100, 0)
day_bosw_df['Non Subscriber %'] = round(day_bosw_df['Non Subscribers'] / day_bosw_df['Amount of Rides'] * 100, 0)
month_bosw_df = day_bosw_df.reset_index()
month_bosw_df.head()
```

```python
month_bosw_df.to_csv("MonthlyWeatherBostonMerged.csv", index=False, header=True)
```
