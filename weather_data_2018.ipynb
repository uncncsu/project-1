```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
#IDs to city names: 4975802 - Portland, OR; 4930956 - Boston, MA; 
bos_data = 'weather_data_BOS_2018.csv'
por_data = 'weather_data_POR_2018.csv'
bdata = pd.read_csv(bos_data)
pdata = pd.read_csv(por_data)
bdata.head()
```

```python
pdata.head()
```

```python
#function to covert temp (in Kelvin) to Farhrenheit
def KelvintoFahr(temp_kelvin):
    converted_temp = (((temp_kelvin - 273.15) * 1.8) + 32)
    return converted_temp
```

```python
#converting Kel to Farh on smaller Boston Data Set
col_name = 'avg_temp_f'
bdata[col_name] = None

for index, row in bdata.iterrows():
  # Convert the Kelvin temperature of the row into Farhrenheit
  farh = KelvintoFahr(row['temp'])
  # Add that value into 'avg_temp_f' column using the index of the row
  bdata.loc[index, col_name] = farh

bdata.head()
```

```python
#converting Kel to Farh on smaller Portland Data Set
col_name = 'avg_temp_f'
pdata[col_name] = None

for index, row in pdata.iterrows():
  # Convert the Kelvin temperature of the row into Farhrenheit
  farh = KelvintoFahr(row['temp'])
  # Add that value into 'avg_temp_f' column using the index of the row
  pdata.loc[index, col_name] = farh

pdata.head()
```

```python
#removed UTC timezone info from dt_iso column
bdata['dt_iso'] = bdata['dt_iso'].astype(str).str[:-9]
pdata['dt_iso'] = pdata['dt_iso'].astype(str).str[:-9]
pdata.head()
```

```python
#There are 8760 hours in a year, so there should be around that many rows.
pdata.count
```

```python
#There are 8760 hours in a year, so there should be around that many rows.
bdata.count
```

```python
#removing duplicates from boston and portland data looking for unique data/time values in dt_iso
bdata = bdata.drop_duplicates(subset='dt_iso', keep='first')
#removing duplicates from portland
pdata = pdata.drop_duplicates(subset='dt_iso', keep='first')
pdata.count
```

```python
#with duplicates removed, appending the portland dataset to boston dataset
wdata = bdata.append([pdata])
wdata.count
```

```python
#removing unneeded columns
columns = ['lat','lon','sea_level','grnd_level','wind_deg',
          'rain_1h','rain_3h','rain_24h','rain_today','snow_1h',
          'snow_3h','snow_24h','snow_today']
wdata.drop(columns, inplace=True, axis=1)
wdata.head()
```

```python
export_csv = wdata.to_csv('weather_data.csv')
```
