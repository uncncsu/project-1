```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
#IDs to city names: 4975802 - Portland, OR; 4930956 - Boston, MA; 
bos_data = 'weather_data_BOS_2018.csv'
por_data = 'weather_data_POR_2018.csv'
bdata = pd.read_csv(bos_data)
pdata = pd.read_csv(por_data)
pdata.head()
```

```python
pdata.head()
```

```python
#function to covert temp (in Kelvin) to Farhrenheit
def KelvintoFahr(temp_kelvin):
    converted_temp = (((temp_kelvin - 273.15) * 1.8) + 32)
    return converted_temp
```

```python
pdata.dtypes
```

```python
#used .apply to call the convert function and add to new column
bdata['avg_temp_f'] = bdata['temp'].apply(KelvintoFahr)
pdata['avg_temp_f'] = pdata['temp'].apply(KelvintoFahr)
pdata.head()
```

```python
#removed UTC timezone info from dt_iso column
bdata['dt_iso'] = bdata['dt_iso'].astype(str).str[:-9]
pdata['dt_iso'] = pdata['dt_iso'].astype(str).str[:-9]
pdata.head()
```

```python
#There are 8760 hours in a year, so there should be around that many rows.
pdata.count
```

```python
#There are 8760 hours in a year, so there should be around that many rows.
bdata.count
```

```python
#removing duplicates from boston and portland data looking for unique data/time values in dt_iso
bdata = bdata.drop_duplicates(subset='dt_iso', keep='first')
#removing duplicates from portland
pdata = pdata.drop_duplicates(subset='dt_iso', keep='first')
pdata.count
```

```python
#convert weather_main to a integer to detemrine mode for day, which will be overall weather pattern
weather ={
    "Clear" : 1,
    "Clouds" : 2,
    "Drizzle": 3,
    "Fog": 4,
    "Haze": 5,
    "Mist": 6,
    "Rain": 7,
    "Snow": 8,
    "Thunderstorm": 9
}
bdata['weather_type'] = bdata['weather_main'].map(weather)
pdata['weather_type'] = pdata['weather_main'].map(weather)
pdata.head()
```

```python
#create unneeded columns
columns = ['city_name','lat','lon','sea_level','grnd_level','wind_deg',
          'rain_1h','rain_3h','rain_24h','rain_today','snow_1h',
          'snow_3h','snow_24h','snow_today', 'clouds_all','weather_id','weather_description']
#remove unneeded columns from boston & portland data
bdata.drop(columns, inplace=True, axis=1)
pdata.drop(columns, inplace=True, axis=1)
pdata.head()
```

```python
#remove the hours from dt_iso so we can group by date
bdata['date'] = bdata['dt_iso'].astype(str).str[:-9]
pdata['date'] = pdata['dt_iso'].astype(str).str[:-9]
pdata.head()
```

```python
bdata.to_csv('hour_weather_bos.csv')
pdata.to_csv('hour_weather_por.csv')
```

```python
#create aggregate data for each city, by day
bdata = bdata.groupby('date').agg({'city_id': 'max',
                                   'pressure': 'mean',
                                   'humidity': 'mean',
                                   'wind_speed': 'mean',
                                   'avg_temp_f': 'mean',
                                   'weather_type': pd.Series.mode
                                             })
bdata.head(25)
```

```python
pdata = pdata.groupby('date').agg({'city_id': 'max',
                                   'pressure': 'mean',
                                   'humidity': 'mean',
                                   'wind_speed': 'mean',
                                   'avg_temp_f': 'mean',
                                   'weather_type': pd.Series.mode
                                             })
pdata.head(50)
```

```python
bdata.to_csv('daily_weather_bos.csv')
pdata.to_csv('daily_weather_por.csv')
```
