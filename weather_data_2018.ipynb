```python
#import libraries
import pandas as pd
import numpy as np
```

```python
# Name of the CSV file
#IDs to city names: 4975802 - Portland, OR; 4930956 - Boston, MA; 
bos_data = 'weather_data_BOS_2018.csv'
por_data = 'weather_data_POR_2018.csv'
bdata = pd.read_csv(bos_data)
pdata = pd.read_csv(por_data)
bdata.head()
```

```python
pdata.head()
```

```python
def KelvintoFahr(temp_kelvin):
    converted_temp = (((273.15 - temp_kelvin) * 1.8) + 32)
    return converted_temp
```

```python
for idx, row in pdata.iterrows():
    print('Index:', idx)
    print(row)
    break
```

```python
col_name = 'Avg_Temp_F'
pdata[col_name] = None

for idx, row in pdata.iterrows():
  # Convert the Fahrenheit temperature of the row into Celsius
  farh = KelvintoFahr(row['temp'])
  # Add that value into 'Celsius' column using the index of the row
  pdata.loc[idx, col_name] = farh

pdata.head()
```

```python
col_name = 'Avg_Temp_F'
bdata[col_name] = None

for idx, row in bdata.iterrows():
  # Convert the Fahrenheit temperature of the row into Celsius
  farh = KelvintoFahr(row['temp'])
  # Add that value into 'Celsius' column using the index of the row
  bdata.loc[idx, col_name] = farh

bdata.head()
```

```python
#removed UTC timezone info from dt_iso column
bdata['dt_iso'] = bdata['dt_iso'].astype(str).str[:-9]
pdata['dt_iso'] = pdata['dt_iso'].astype(str).str[:-9]
pdata.head()
```

```python
#There are 8760 hours in a year, so there should be around that many rows.
pdata.count
```

```python
#removing duplicates from boston data sorting on ID and looking for unique data/time values in dt_iso
bdata = bdata.sort_values('city_id', ascending=False)
bdata = bdata.drop_duplicates(subset='dt_iso', keep='first')
#removing duplicates from portland
pdata = pdata.sort_values('city_id', ascending=False)
pdata = pdata.drop_duplicates(subset='dt_iso', keep='first')
pdata.count
```

```python
#with duplicates removed, appending the portland dataset to boston dataset
wdata = bdata.append([pdata])
wdata.count
```
